# working with date

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# 1. extract year
date['date_year'] = date['date'].dt.year

date.head()
```

```python
# 1. extract month
date['date_month'] = date['date'].dt.month

date.head()
```

```python
# 1. extract day
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# 1. extract month_name
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

```python
# 1. extract day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# 1. extract day of week name
date['date_dow_name'] = date['date'].dt.day_name()

date.head()
```

```python
# 1. extract day of week weekday
date['date_dow_weekday'] = date['date'].dt.weekday

date.head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)

date.head()
```

```python
# 1. extract week of the year
date['date_week'] = date['date'].dt.week

date.head()
```

```python
# 1. extract quater
date['quater'] = date['date'].dt.quarter

date.head()
```

```python
# 1. extract Semetster
date['Semester'] = np.where(date['quater'].isin([1,2]),1,2)

date.head()
```

# Extract Time elapsed between dates

```python
import datetime

today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# month passed

np.round((today - date['date'])/np.timedelta64(1,'M'))
```

```python
import pandas as pd
import pytz

index = pd.date_range('20220101 21:55', freq='31536000S', periods=20)
df = pd.DataFrame(1, index=index, columns=['X'])
print(df)
#                      X
# 2014-01-01 21:55:00  1
# 2014-01-01 21:55:15  1
# 2014-01-01 21:55:30  1
# 2014-01-01 21:55:45  1
# 2014-01-01 21:56:00  1

# [5 rows x 1 columns]
print(df.index)
# <class 'pandas.tseries.index.DatetimeIndex'>
# [2014-01-01 21:55:00, ..., 2014-01-01 21:56:00]
# Length: 5, Freq: 15S, Timezone: None

eastern = pytz.timezone('US/Eastern')
df.index = df.index.tz_localize(pytz.utc).tz_convert(eastern)
print(df)
#                            X
# 2014-01-01 16:55:00-05:00  1
# 2014-01-01 16:55:15-05:00  1
# 2014-01-01 16:55:30-05:00  1
# 2014-01-01 16:55:45-05:00  1
# 2014-01-01 16:56:00-05:00  1

# [5 rows x 1 columns]

print(df.index)
```
