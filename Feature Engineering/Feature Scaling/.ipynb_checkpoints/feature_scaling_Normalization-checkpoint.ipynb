```python
df = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])

df.columns = ['Class_label','Alcohol', 'Malic_acid']
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('Class_label', axis=1),
                  df.Class_label, test_size=0.2,
                  random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

# MinMaxScaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
np.round(X_train.describe(),2)
```

```python
np.round(X_train_scaled.describe(),2)
```

```python
fig, (ax,ax1) = plt.subplots(ncols=2,figsize= (12,5))

ax.scatter(X_train.Alcohol,X_train.Malic_acid,c=y_train)
ax.set_title('Before Normalization')

ax1.scatter(X_train_scaled.Alcohol,X_train_scaled.Malic_acid,c=y_train)
ax1.set_title('Afternoon  Normalization')
plt.show()
```

```python
fig, (ax,ax1) = plt.subplots(ncols=2,figsize= (12,5))

sns.kdeplot(X_train.Alcohol, ax = ax)
sns.kdeplot(X_train.Malic_acid, ax = ax)
ax.set_title('Before Normalization')

sns.kdeplot(X_train_scaled.Alcohol, ax = ax1)
sns.kdeplot(X_train_scaled.Malic_acid, ax = ax1)
ax.set_title('Before Normalization')
plt.show()
```

# Mean Normalization

```python
def mean_naormalization(df,X):
    for i in df.columns:
        X_mean_naormalization = ((df[X]-df[X].mean())/(df[X].max()-df[X].min()))
    
    return X_mean_naormalization
```

```python
X_train_scaled = pd.DataFrame({X_train.columns[0]:mean_naormalization(X_train,X_train.columns[0]),X_train.columns[1]:mean_naormalization(X_train,X_train.columns[1])})
X_test_scaled = pd.DataFrame({X_test.columns[0]:mean_naormalization(X_test,X_test.columns[0]),X_test.columns[1]:mean_naormalization(X_test,X_test.columns[1])})
```
