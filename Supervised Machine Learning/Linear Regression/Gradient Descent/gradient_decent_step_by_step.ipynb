```python
from sklearn.datasets import make_regression

import numpy as np
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# apply ols method
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='r')
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = b
y_pred = ((78.35 * X) + 100).reshape(4)
y_pred.ravel()
```

```python
plt.scatter(X,y)
# OlS plot
plt.plot(X,reg.predict(X),color='r',label = 'OLS')
# m=78.53,b=0 plot
plt.plot(X,y_pred,color='b',label='b = 0')

plt.legend()
plt.show()
```

```python
m= 78.53
b=100

loss_slope = -2 * np.sum(y - m * X.ravel() - b) 
loss_slope
```

```python
X
```

```python
# Lets take learning rate = 0.1
lr = 0.1

step_size = loss_slope*lr

step_size
```

```python
b_new = b - step_size
b_new
```

```python
y_pred1 = ((78.35 * X) + b_new).reshape(4)

plt.scatter(X,y)
# OlS plot
plt.plot(X,reg.predict(X),color='r',label = 'OLS')
# m=78.53,b=0 plot
plt.plot(X,y_pred1,color='g',label='b = {}'.format(b_new))
plt.plot(X,y_pred,color='b',label='b = 0')
plt.legend()
plt.show()
```

```python
# iteration 2
loss_slope = -2 * np.sum(y - m * X.ravel() - b_new) 
loss_slope
```

```python
step_size = loss_slope*lr

step_size
```

```python
b_new1 = b_new - step_size
b_new1
```

```python
y_pred2 = ((78.35 * X) + b_new1).reshape(4)

plt.scatter(X,y)
# OlS plot
plt.plot(X,reg.predict(X),color='r',label = 'OLS')
# m=78.53,b=0 plot
plt.plot(X,y_pred1,color='g',label='b = {}'.format(b_new))
plt.plot(X,y_pred2,color='k',label='b = {}'.format(b_new1))
plt.plot(X,y_pred,color='b',label='b = 0')
plt.legend()
plt.show()
```

```python
# iteration 3
loss_slope = -2 * np.sum(y - m * X.ravel() - b_new1) 
loss_slope
```

```python
step_size = loss_slope*lr

step_size
```

```python
b_new2 = b_new1 - step_size
b_new2
```

```python
y_pred3 = ((78.35 * X) + b_new2).reshape(4)

plt.scatter(X,y)
# OlS plot
plt.plot(X,reg.predict(X),color='r',label = 'OLS')
# m=78.53,b=0 plot
plt.plot(X,y_pred1,color='g',label='b = {}'.format(b_new))
plt.plot(X,y_pred2,color='k',label='b = {}'.format(b_new1))
plt.plot(X,y_pred3,label='b = {}'.format(b_new2))
plt.plot(X,y_pred,color='b',label='b = 0')
plt.legend()
plt.show()
```
