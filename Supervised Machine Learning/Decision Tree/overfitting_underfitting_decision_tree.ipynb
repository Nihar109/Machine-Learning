```python
import numpy as np
import pandas as pd
 
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
df.shape
```

```python
def analyzer(max_depth):
    df = pd.read_csv('Social_Network_Ads.csv')
    
    X = df.iloc[:,2:-1].values
    y = df.iloc[:,-1].values
    
    clf = DecisionTreeClassifier(max_depth=max_depth)
    clf.fit(X,y)
    
    a = np.arange(X[:,0].min()-1,X[:,0].max()+1,0.1)
    b = np.arange(X[:,1].min()-1,X[:,1].max()+1,100)
    
    XX,YY = np.meshgrid(a,b)
    
    X_input = np.array([XX.ravel(),YY.ravel()]).T
    y_input = clf.predict(X_input)
    
    plt.contourf(XX,YY,y_input.reshape(XX.shape),alpha=0.5)
    plt.scatter(X[:,0],X[:,1],c=y)
```

```python
# Overfitting
analyzer(max_depth=None)
```

```python
# underfitting
analyzer(1)
```

```python
# Optimum max depth
analyzer(4)
```
