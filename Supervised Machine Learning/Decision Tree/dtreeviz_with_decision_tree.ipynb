```python
# Univariate Regressor

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from dtreeviz.trees import *

df_cars = pd.read_csv("cars.csv")
X, y = df_cars[['wt']], df_cars['mpg']

dt = DecisionTreeRegressor(max_depth=3, criterion="mae")
dt.fit(X,y)
```

```python
plot_tree(dt)
```

```python
viz_rmodel = dtreeviz.model(dt, X, y,
                           feature_names=['wt'],
                           target_name='mpg')

viz_rmodel.rtree_feature_space()
```

```python
# 3_D Regressor

X = df_cars[["wt", "disp"]]
y = df_cars['mpg']
dtr_bivar_3d = DecisionTreeRegressor(max_depth=3, criterion="absolute_error")
dtr_bivar_3d.fit(X.values, y.values)
```

```python
viz_rmodel = dtreeviz.model(dtr_bivar_3d, X, y,
                           feature_names=["wt", "disp"],
                           target_name='mpg')
viz_rmodel.view(scale=2)
```

```python
viz_rmodel.rtree_feature_space3D(fontsize=10,
                        elev=30, azim=20, 
                        show={'splits','title'},
                        colors={'tesselation_alpha':.5})
```
