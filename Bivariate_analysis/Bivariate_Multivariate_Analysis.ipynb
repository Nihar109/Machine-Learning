# 4. Distplot (Numerical - Categorical)

```python
sns.displot(x='Age',data=titanic,kde=True)
```

```python
fig, ax = plt.subplots()
sns.kdeplot(titanic[titanic['Survived']==0]['Age'],ax=ax)
sns.kdeplot(titanic[titanic['Survived']==1]['Age'],ax=ax)
plt.show()
```

# 5. Heatmap (Categorical - Categorical)

```python
titanic.sample()
```

```python
cat_dat = pd.crosstab(titanic.Pclass,titanic.Survived)
cat_dat
```

```python
sns.heatmap(cat_dat)
```

```python
titanic.groupby('Pclass')['Survived'].mean()
```

```python
titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
titanic.groupby('Sex')['Survived'].mean()*100
```

```python
(titanic.groupby('Sex')['Survived'].mean()*100).plot(kind='bar')
```

```python
(titanic.groupby('Embarked')['Survived'].mean()*100)
```

```python
(titanic.groupby('Embarked')['Survived'].mean()*100).plot(kind='bar')
```
